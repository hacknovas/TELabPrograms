```python
dframe2=dframe
dframe2['M2']=dframe2['M2'].fillna(dframe2['M2'].mean())
dframe2['M3']=dframe2['M3'].fillna(dframe2['M3'].mean())
dframe2
```

```python
dframe2=dframe
array1=dframe2['M2']
#print(array1)
array1.sort_values()

dframe2['M2']=dframe2['M2'].fillna(array1.median())
#dframe2['M3']=dframe2['M3'].fillna(array2.median())
dframe2
```

```python
#using standard deviation
#standard deviation=sqrt((sum(xi-mean)^2)/N) 
dframe2=dframe
dframe2['M2']=dframe2['M2'].fillna(dframe2['M2'].std())
dframe2
#dframe2['M2'].value_counts()
#dframe2['M3']=dframe2['M3'].fillna(array2.median())
#dframe2
```

```python
#minimum value
dframe2=dframe
dframe2['M2']=dframe2['M2'].fillna(dframe2['M2'].min())
dframe2
```

```python
#maximum value
dframe2=dframe
dframe2['M3']=dframe2['M3'].fillna(dframe2['M3'].max())
dframe2
```
